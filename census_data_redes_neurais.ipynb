```python
import pickle

with open('census.pkl', 'rb') as f:
    X_census_treinamento, y_census_treinamento, X_census_teste, y_census_teste = pickle.load(f)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
censusRedeNeural = MLPClassifier(verbose=True, max_iter=1000, tol=0.0000100, hidden_layer_sizes=(55,55))

censusRedeNeural.fit(X_census_treinamento, y_census_treinamento)
```

```python
previsoes = censusRedeNeural.predict(X_census_teste)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_census_teste, previsoes)
```

```python
from yellowbrick.classifier import ConfusionMatrix

cm = ConfusionMatrix(censusRedeNeural)
cm.fit(X_census_treinamento, y_census_treinamento)
cm.score(X_census_teste, y_census_teste)
```

```python
print(classification_report(y_census_teste, previsoes))
```
