```python
import pickle

with open('census.pkl', 'rb') as f:
    X_census_treinamento, y_census_treinamento, X_census_teste, y_census_teste = pickle.load(f)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
censuLogisticRegression = LogisticRegression(random_state=1)

censuLogisticRegression.fit(X_census_treinamento, y_census_treinamento);
```

```python
previsoes = censuLogisticRegression.predict(X_census_teste)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_census_teste, previsoes)
```

```python
from yellowbrick.classifier import ConfusionMatrix

cm = ConfusionMatrix(censuLogisticRegression)
cm.fit(X_census_treinamento, y_census_treinamento)
cm.score(X_census_teste, y_census_teste)
```

```python
print(classification_report(y_census_teste, previsoes))
```
